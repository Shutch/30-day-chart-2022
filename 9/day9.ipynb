# Distributions - Statistical

Guam household income statistics have changed significantly over time, with wages rising, population rising, and USD inflation rising as well.

## Data

The [Guam Statistical Yearbook](https://bsp.guam.gov/guam-statistical-yearbook-2/) table 13-14 contains per household income from 2001, 2003, 2005, 2008 and 2010. I manually moved that data into a CSV file located in this folder. I also created an additional CSV containing some statistics such as median and mean household income for each year.

```python
import pandas as pd

df = pd.read_csv("household_income_rebucketed.csv")
df_stats = pd.read_csv("household_income_stats.csv", index_col='Statistics')

print(df.head())
print(df_stats.head())
```

```python
import matplotlib.ticker as mtick

start_year = '2001'
end_year = '2010'

median_start = int(df_stats.loc['Median Household Income', start_year])
median_end = int(df_stats.loc['Median Household Income', end_year])
mean_start = int(df_stats.loc['Mean Household Income', start_year])
mean_end = int(df_stats.loc['Mean Household Income', end_year])
median_change = (median_end - median_start) / median_start * 100
mean_change = (mean_end - mean_start) / mean_start * 100
inflation = float(df_stats.loc['Inflation', end_year]) * 100

f, ax = plt.subplots(figsize=(4, 8), facecolor='white')

plt.bar(x=[0, 1, 2],
        height=[median_change, mean_change, inflation],
        color=['mediumpurple', 'sandybrown', 'steelblue'],
        tick_label=['median', 'mean', 'inflation'])

sns.despine(left=True)
ax.set_ylabel("Percent Change")
plt.grid(axis='x', visible=False)
ax.yaxis.set_major_formatter(mtick.PercentFormatter())
ax.set_xticklabels(['Median Income', 'Mean Income', 'Cumulative Inflation'],
                   rotation=45,
                   fontsize=12,
                   ha='right')
plt.title("Percent Change in Inflation, Median and \nMean Household Income From 2001 to 2010", fontsize=20)

plt.show()
```
