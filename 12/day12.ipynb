# Distributions - The Economist

According to the Economist [here](https://design-system.economist.com/documents/CHARTstyleguide_20170505.pdf) and [here](https://medium.economist.com/charting-new-territory-7f5afb293270), the web chart style used is minimalist with black font, blue data, and red highlights. The charts are small and simple, with captions and subtitles conveying additional information.

## Data

The Guam Statistical Yearbook table 6-01 contains student population by school type from 2016-2021 I manually moved that data into a CSV file located in this folder.

```python
from matplotlib.colors import ListedColormap

# setting up colors
red =   "#E3120B"
black = "#0C0C0C"
text = "#5a5d5e"
grid_line = "#B7C6CF"
c1 = "#106ea0"
c2 = "#32c0d2"
c3 = "#e0b165"
c4 = "#00969e"

stacked_cmap = ListedColormap([c1, c2, c3, c4])
stacked_cmap
```

```python
import pandas as pd

df = pd.read_csv('guam_education_data.csv')

df
```

```python
grades_df = df.groupby(by='School').sum().transpose().sort_index()
grades_df
```

```python
# white lines for plotting between area plot sections
grades = ["Kindergarten to 5th Grade",
          "6th - 8th Grade",
          "9th - 12th Grade"]
          #"High School Graduates"]
grade_lines = ['_5-6 Line_', '_8-9 Line_', '_12 Line_']
grades_df[grade_lines[0]] = grades_df[grades[0]]
grades_df[grade_lines[1]] = grades_df[grades[0]] + grades_df[grades[1]]
grades_df[grade_lines[2]] = grades_df[grades[0]] + grades_df[grades[1]] + grades_df[grades[2]]
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle

fig, ax = plt.subplots(figsize=(12, 6), facecolor='white')

years = list(range(2016, 2022))
grades_df.plot.area(y=grades,
                    ax=ax,
                    colormap=stacked_cmap)
grades_df.plot(y=grade_lines,
               ax=ax,
               color='w',
               linewidth=2)

ax.set_axisbelow(True)
ax.grid(color=grid_line, linestyle='-', linewidth=2, axis='y')
ax.spines['left'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.yaxis.set_label_position("right")
ax.yaxis.tick_right()
ax.yaxis.set_tick_params(length=0)

plt.suptitle("Guam's Declining Student Population",
             x=0.125,
             y=1.02,
             ha='left',
             fontsize=18,
             fontweight='bold')
plt.title("Number of public and private students by school type",
          x=0,
          y=1.06,
          ha='left')
plt.legend(loc='upper left',
           bbox_to_anchor=(0.0, 1.08),
           ncol = 3,
           fontsize=14,
           frameon=False)
plt.figtext(x=0.12,
            y=0.04,
            s="Source: Guam Statistical Yearbook 2020, Table 6-01",
            ha="left",
            fontsize=12,
            color=text)

# Red bar and block in top left
r = Rectangle(xy=(0.0, 1.22),
              width=0.05,
              height=0.03,
              color=red,
              transform=ax.transAxes,
              clip_on=False)
ax.add_patch(r)

r = Rectangle(xy=(0.0, 1.248),
              width=1.05,
              height=0.0025,
              color=red,
              transform=ax.transAxes,
              clip_on=False)
ax.add_patch(r)

plt.show()
```
